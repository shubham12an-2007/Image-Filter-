```python
import numpy as np 
from PIL import Image
```

```python
img  = Image.open("goku.png").convert("RGB")

img_array = np.array(img , dtype=float)
```

```python
height , width , channels = img_array.shape

output_array = np.zeros_like(img_array)

kernel = np.array([[1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9]])

print("Filtering of image has been started")
```

```python
for c in range(channels):
  for y in range(1 , height  - 1):
    for x in range(1,width -1 ):
      neighborhood = img_array[y-1:y+2, x-1:x+2, c]

      new_value  = np.sum(neighborhood * kernel)

      output_array[y, x, c] = new_value
    
#  border pixels copied from the original array
output_array[0, :, :] = img_array[0, :, :]
output_array[-1, :, :] = img_array[-1, :, :]
output_array[:, 0, :] = img_array[:, 0, :]
output_array[:, -1, :] = img_array[:, -1, :]


# convert final output in to the image 
output_array = np.clip(output_array ,  0 , 255).astype(np.uint8)

finalImage = Image.fromarray(output_array)
finalImage.save("smoothed_goku.png")

print("Image smoothed successfully")
```
